```python
%matplotlib inline
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import statsmodels.api as sm

import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
pylab.rcParams.update(params)
plt.xkcd()
```

```python
df = pd.DataFrame.from_csv(r'data\student_height_weight.csv',index_col=None)
```

```python
df.head()
```

```python
df1 = df[(df.height<=80)&(df.weight<500)]

fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(1, 1, 1)
plt.scatter(df1.height,df1.weight)
ax.set_title('Height Vs Weight')
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
```

```python
X = df1.height.values.tolist()
X = sm.add_constant(X)
y = df1.weight
model = sm.OLS(y, X)
results = model.fit()
pred_y = results.predict(X)
print(results.summary())
```

```python
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(1, 1, 1)
plt.scatter(df1.height,df1.weight)
ax.set_title('Height Vs Weight')
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.plot(df1.height,pred_y,color='red')
```
